### Kmeans

```python
import numpy as np
from scipy.spatial.distance import cdist
import matplotlib.pyplot as plt
```

```python
# khởi tạo ban đâu các trung tâm của K cluster
def init_K_centroid(k, X):
    return X[np.random.choice(np.shape(X)[0], k, replace=False)]


# phân các điểm dữ liệu vào các cluster
def assign_label(X, centroids):
    return np.argmin(cdist(X, centroids), axis=1)


# tính lại trung tâm
def create_K_centroid(centroids, X, k, label):
    centroids_after = np.zeros(np.shape(centroids))
    for i in range(k):
        centroids_after[i] = np.mean(X[np.where(label == i)], axis=0)
    return centroids_after

# điều kiện dừng khi các trung tâm có khoảng cách nhỏ trước so với sau
def condition(centroids_before, centroids_after, esp=1e-2):
    bal = np.linalg.norm(centroids_after - centroids_before)
    return bal < esp

# thuật toán kmeans thông thường, trả về k centers và các label của từng điểm dữ liệu
def K_Mean_Clustering(k, X, random_state=42):
    np.random.seed(random_state)
    centroids_before = init_K_centroid(k, X)
    label = assign_label(X, centroids_before)
    while True:
        centroids_after = create_K_centroid(centroids_before, X, k, label)
        label = assign_label(X, centroids_after)
        if condition(centroids_before, centroids_after):
            break
        centroids_before = np.copy(centroids_after)
    return centroids_after, label
```

```python
means = [[2, 2], [8, 3], [3, 6]]
cov = [[2, 0], [0, 2]]
N = 300
Xi = []

for i in range(3):
    x = np.random.multivariate_normal(means[i], cov, N)
    Xi.append(x)
    
X = np.array([y for x in Xi for y in x])

centroids, label = K_Mean_Clustering(3, X)
print(centroids)
print(label)
```

```python
uni_label = np.unique(label)

plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.title('Original')
for x in Xi:
    plt.scatter(x[:, 0], x[:, 1])

plt.subplot(1,2,2)
plt.title('Kmean')
for l in uni_label:
    x = X[np.where(label == l)]
    plt.scatter(x[:, 0], x[:,1])
plt.scatter(centroids[:, 0], centroids[:, 1], color='black')

plt.tight_layout()
plt.show()
```

#### Streaming kmeans tương tự như kmeans cluster, tuy nhiên dữ liệu không thể nạp hết vào bộ nhớ <br>
Các bước thuật toán (ở đây ta chỉ cài lại phiên bản đơn giản, do thuật toán còn có nhiều tham số để phù hợp yêu câu bộ nhớ, ...) : <br>
    1,Chia dữ liệu X ra thành nhiều phần X1, X2, ... Xl<br>
    2,Thực hiện thuật toán kmeans cho từng Xi<br>
    3,Khi đó ta có l*k centers, thực hiện kmeans cho các center đó

```python
def Streaming_K_means(X,k=3,l=3,random_state=42):
    np.random.seed(random_state)
    xs = np.split(X,l) # chia dữ liệu thành l phần
    # thực hiện kmean cho từng phần để có center
    centroids, i = {'center':[], 'label':[]}, 0
    for x in xs:
        cens, labels = K_Mean_Clustering(k, x, random_state)
        labels += i
        i += k
        centroids['center'].extend(cens.tolist())
        centroids['label'].extend(labels.tolist())
    # kmean cho các center
    result_cs, ls = K_Mean_Clustering(k, np.array(centroids['center']), random_state)
    # lấy kết quả nhãn sau khi chia cluster cho các center
    result_label = ls[centroids['label']]
    return centroids, result_cs, result_label
```

```python
cs, rcs, rl = Streaming_K_means(X)
cs['center'] = np.array(cs['center'])
cs['label'] = np.array(cs['label'])

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.title('Original')
for x in Xi:
    plt.scatter(x[:, 0], x[:, 1])

plt.subplot(1,3,2)
ul = np.unique(cs['label'])
for i in ul:
    x = X[np.where(cs['label'] == i)]
    plt.scatter(x[:, 0], x[:, 1])
    
plt.scatter(cs['center'][:, 0], cs['center'][:, 1], color='black')

plt.subplot(1,3,3)
plt.title('Streaming Kmeans')
ul = np.unique(rl)
for i in ul:
    x = X[np.where(rl == i)]
    plt.scatter(x[:, 0], x[:, 1])
    
plt.scatter(rcs[:, 0], rcs[:, 1], color='black')

plt.show()
```
