Merge k danh sách đã được sắp xếp 

Input:
[
  1->4->7,
  1->3->4,
  2->6
]
Output: 1->1->2->3->4->4->6->7

Implement với tối thiểu 2 cách khác nhau. Đánh giá độ phức tạp của từng phương pháp

# Cách 1 : sử dụng giống như merge sort

Tại mỗi vòng lặp số ta merge sao cho số lượng danh sách còn lại một nửa<br>
Số vòng lặp ta cần thực hiện là log(k)

```python
k_lists = [[1,4,7], [1, 3, 4], [2,6]]
print(k_lists)
```

#### Hàm merge 2 danh sách đã sắp xếp

```python
# độ phức tạp O(tổng số phần tử 2 danh sách)
def merge(l1, l2):
    l_merge, len1, len2, i, j = [], len(l1), len(l2), 0, 0
    while i < len1 and j < len2:
        if l1[i] <= l2[j]:
            l_merge.append(l1[i])
            i += 1
        else : 
            l_merge.append(l2[j])
            j += 1
    while i < len1:
        l_merge.append(l1[i])
        i += 1
    while j < len2:
        l_merge.append(l2[j])
        j += 1
    return l_merge
```

```python
print(merge([1,4,7], [1, 3, 4]))
```

```python
def merges(k_lists):
    k = len(k_lists)
    l_merge = list(k_lists).copy()
    # lặp cho đến khi mà tắt cả danh sách đã được hợp làm 1
    while k > 1: # số lần lặp là log(k)
        # độ phức tạp tại mỗi vòng lặp for là số phần tử của tất cả danh sách
        for i in range(k//2):
            l_merge[i] = merge(l_merge[i], l_merge[k - i - 1])
        k=k//2 + k%2
    # độ phức tạp O(n*log(k)), trong đó n là sô phần tử của k danh sách
    return l_merge[0] 
```

```python
print(merges(k_lists))
```

# Cách 2 : sử dụng min heap

tương tự như thuật heap sort nhưng heap luôn chỉ có tối đa k phần tử

```python
import heapq
```

```python
def merge_heap(k_lists):
    k = len(list(k_lists))
    # danh sách các phần tử đầu tiên của k mảng, heap ban đâu có k phần tử
    heap = [l[0] for l in k_lists if len(l) > 0]
    # mảng chứa tất cả các phần tử còn lại của danh sách
    combine_list = [y for x in k_lists if len(x) > 1 for y in x[1:]]
    l_merge = []
    for i in combine_list:
        # nếu root heap lớn hơn phần tử đang xét thì thêm nó vào kết quả, độ phức tạp O(1)
        # ngược lại đẩy root vào kết quả và chèn i vào heap , độ phức tạp O(log(k))
        root = heapq.nsmallest(1, heap)[0]
        if i < root : l_merge.append(i)
        else :
            l_merge.append(root)
            heapq.heapreplace(heap, i)
    # độ phức tạp O(n*log(k)), n là tổng số phần tử k danh sách
    l_merge.extend(heapq.nsmallest(k, heap))
    return l_merge
```

```python
print(merge_heap(k_lists))
```
